```python
import pandas as pd

imdb_df = pd.read_csv("data/imdb_data.csv")
df1 = pd.read_csv("metacritic_shows1.csv")
df2 = pd.read_csv("metacritic_shows2.csv")
df1 = df1.drop_duplicates()
df2 = df2.drop_duplicates()

print(df1.shape, df2.shape)
```

```python
merged = pd.merge(df1, df2, on='id', how='outer', suffixes=('_df1', '_df2'))

# Find rows that exist only in one dataframe
only_in_df1 = merged[merged['title_df2'].isnull()].drop(columns=[col for col in merged.columns if '_df2' in col])
only_in_df2 = merged[merged['title_df1'].isnull()].drop(columns=[col for col in merged.columns if '_df1' in col])

print(only_in_df1.shape, only_in_df2.shape)
print("Rows present only in df1:")
display(only_in_df1)

print("\nRows present only in df2:")
display(only_in_df2)
```

```python
df3 = pd.concat([df1, df2])
df3 = df3.drop_duplicates()
filtered_df = df3[df3['release'] >= 2000]
filtered_df.to_csv('data/metacritic_data.csv', index=False)
```
